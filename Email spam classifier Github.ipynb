```python
import numpy as np
import pandas as pd
import os
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
folder='email/'
```

```python
files=[]
files=os.listdir(folder)
```

```python
len(files)
```

```python
files[1:10]
```

```python
emails=[folder+file for file in files]
```

```python
emails[1:10]
```

```python
words=[]
for email in emails:
    f=open(email,encoding='latin-1')
    blob=f.read()
    words+=blob.split(" ")
```

```python
blob
```

```python
len(words)
```

```python
for i in range(len(words)):
    if not words[i].isalpha():
        words[i]=""
```

```python
words_dict=Counter(words)
```

```python
words_dict
```

```python
len(words_dict)
```

```python
del words_dict[""]
```

```python
words_dict
```

```python
len(words_dict)
```

```python
words_dict=words_dict.most_common(3000)
```

```python
len(words_dict)
```

```python
features=[]
target=[]
for email in emails:
    f=open(email,encoding='latin-1')
    blob=f.read().split(" ")
    data=[]
    for i in words_dict:
        data.append(blob.count(i[0]))
    features.append(data)
    if 'spam' in email:
        target.append(0)
    if 'ham' in email:
        target.append(1)
```

```python
len(features)
```

```python
len(target)
```

```python
features
```

```python
X=np.array(features)
```

```python
X.shape
```

```python
y=np.array(target)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
```

```python
classifier.fit(X_train,y_train)
```

```python
def converter(sample):
    data=[]
    for i in words_dict:
        data.append(sample.split(" ").count(i[0]))
    emailInput=np.array(data)
    
    return emailInput.reshape(1,3000)
```

```python
y_pred=classifier.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
